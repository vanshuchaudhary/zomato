```python
df1 = df.copy()
df1.head()
```

```python
df1["Cuisines"] = df1["Cuisines"].astype(str)
```

```python
#mode imputation
mode_value =  df1["Cuisines"].mode()[0]
mode_value 
```

```python
df1["Cuisines"] = df["Cuisines"].fillna(mode_value)
df1["Cuisines"].isnull().sum()
```

```python
#mode imputation - has_table_booking 

df1["Has_Table_booking"]=df1["Has_Table_booking"].astype("str")
mode_val = df1["Has_Table_booking"].mode()[0]
df1["Has_Table_booking"]=df1["Has_Table_booking"].replace("3",mode_val)
df1["Has_Table_booking"].unique()
```

```python
mode_value1 = df1["Locality"].mode()[0]
df1["Locality"] = df1["Locality"].astype(str)
df1["Locality"] = df1["Locality"].replace('51',mode_value1)
df1["Locality"].unique()
```

```python
#Has_Online_delivery
df1["Has_Online_delivery"]=df1["Has_Online_delivery"].astype("str")
mode_val = df1["Has_Online_delivery"].mode()[0]
df1["Has_Online_delivery"]=df1["Has_Online_delivery"].replace("3",mode_val)
df1["Has_Online_delivery"].unique()
```

```python
#Is_delivering_now
df1["Is_delivering_now"]=df1["Is_delivering_now"].astype("str")
mode_val = df1["Is_delivering_now"].mode()[0]
df1["Is_delivering_now"]=df1["Is_delivering_now"].replace("3",mode_val)
df1["Is_delivering_now"].unique()
```

```python
#dropping duplicated value 
df1[df1.duplicated(["Restaurant_Name" ,"Restaurant_ID"])].shape
```

```python
df1 = df1.drop_duplicates(subset=["Restaurant_Name" ,"Restaurant_ID"] ,keep="first")
df1.head()
```

```python
#What is the distribution of restaurants across different price ranges?
price_across = df1["Price_range"].value_counts()
price_across

plt.figure(figsize=(10,10))
plt.bar(price_across.index ,price_across.values ,color = "red")
plt.xticks([1,2,3,4])
plt.xlabel("price_range[1 = lower class restaurants ,4 = high class restaurants]")
plt.ylabel("no.of restaurants")
plt.title("Distribution of restaurants across different price ranges")
```

```python
#What is the distribution of aggregate ratings?
aggregate_ratings = df1["Aggregate_rating"].value_counts()
aggregate_ratings

colors = []
for rating in aggregate_ratings.index:
    if rating == 0:
        colors.append("violet")
    elif rating < 1:
        colors.append("green")  
    elif rating == 1:
        colors.append("blue")
    elif rating < 2:
        colors.append("black")  
    elif rating == 2:
        colors.append("orange")
    elif rating < 3:
        colors.append("yellow") 
    elif rating == 3:
        colors.append("skyblue")
    elif rating < 4: 
        colors.append("purple") 
    elif rating == 4:
        colors.append("teal")
    elif rating < 5: 
        colors.append("lime")
    elif rating == 5:
        colors.append("pink") 
    else:  #
        colors.append("pink")

           
plt.figure(figsize=(7,8))
plt.barh(aggregate_ratings.index ,aggregate_ratings.values ,color = colors,edgecolor = "black")
plt.xlabel("Aggregate_rating")
plt.ylabel("count")
plt.title("Distribution of aggregate ratings")
plt.tight_layout()
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
